```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
import scipy
```

```python
df = pd.read_csv("IRIS.csv")
```

```python
df.sample(6)
```

```python
df[df['species'] == 'Iris-setosa']
```

```python
df.corr()
```

```python
x = df['petal_width']
y = df['petal_length']

slope, intercept, r, p, stderr = scipy.stats.linregress(x, y)
slope, intercept
```

```python
line = f'Regression line: y={intercept:.2f}+{slope:.2f}x'
line
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, linewidth=0, marker='o', label='Data points')
ax.plot(x, intercept + slope * x, label=line, color='purple')
ax.set_xlabel('PETAL WIDTH')
ax.set_ylabel('PETAL LENGTH')
ax.set_title('PETAL WIDTH VS LENGTH OF 150 IRIS FLOWERS')
ax.legend(facecolor='white')
plt.show()
```

```python
# Multiple Regression in StatsModels
import statsmodels.api as sm
x = df[['sepal_length','petal_width']]
predictors = sm.add_constant(x)
model = sm.OLS(y, predictors).fit()
model.summary()
```
